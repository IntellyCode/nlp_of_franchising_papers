```python
import pickle

with open('./output/store/corpus.pkl', 'rb') as f:
    corpus, dictionary = pickle.load(f)
```

```python
import os
directory = "./data/All Papers"
pdf_files = [f for f in os.listdir(directory) if f.lower().endswith('.pdf')]
pdf_files.sort() 
pdf_files
```

```python
from gensim import models

lda_directory = os.path.join('./output', 'n_topics_12_min/model')
lda_model = models.LdaModel.load(lda_directory)
num_topics = lda_model.num_topics
topic_distributions = [lda_model[doc] for doc in corpus]
print(topic_distributions[0:100])
```

```python
best_docs = {i: (None, 0) for i in range(num_topics)}

for doc_index, topics in enumerate(topic_distributions):
    for topic_id, prob in topics:
        if prob > best_docs[topic_id][1]: 
            best_docs[topic_id] = (doc_index, prob)

for topic_id, (doc_index, prob) in best_docs.items():
    print(f"Topic {topic_id+1}: Best matching document index {pdf_files[doc_index]} with probability {prob:.4f}")
```
